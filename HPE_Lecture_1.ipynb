```
class Fibo:
    def __iter__(self):
        self.a = -1
        self.b = 1
        self.c = True
        return self

    def __next__(self):
        self.c = self.a + self.b
        self.a = self.b
        self.b = self.c
        return self.c

```

```
series = Fibo()
prop = iter(series)
next(prop)
```

```
xyz = []
for _ in range(10):
    xyz.append(next(prop))

print(xyz)
```

```
import pandas as pd
s = pd.Series(xyz)
```

```
print(s)
```

```
df = {
    'fibo':s,
    'random':range(10)
}
dFrame = pd.DataFrame(df)
```
